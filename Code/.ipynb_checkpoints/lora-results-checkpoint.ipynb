```python
try:
    runs = api.runs(project_name)
    print(f"Found {len(runs)} runs in project {project_name}")
except Exception as e:
    print(f"Error accessing W&B project: {e}")
    exit(1)
```

### Reconstruct results list

```python
results = []
```

```python
results = []
for run in runs:
    # Extract hyperparameters and metrics
    rank = run.config.get("rank")
    lr = run.config.get("learning_rate")
    modules = run.config.get("target_modules")
    train_loss = run.summary.get("train_loss", float('inf'))
    eval_loss = run.summary.get("eval_loss", float('inf'))  # Assuming eval_loss is val_loss
    trainable_params = run.summary.get("trainable_params", 0)
    squad_exact_match = run.summary.get("squad_exact_match_greedy_{}", 0)
    squad_f1 = run.summary.get("squad_f1_greedy_{}", 0)
    
    # Map modules to short names for clarity
    module_short_name = {
        str(["attn.c_attn", "attn.c_proj"]): "attn",
        str(["attn.c_attn", "attn.c_proj", "mlp.c_fc", "mlp.c_proj"]): "attn_mlp",
        str(["q_proj", "v_proj", "attn.c_attn", "attn.c_proj"]): "attn_qv"
    }.get(str(modules), "custom")
    
    # Only include runs with valid metrics
    if rank is not None and lr is not None and modules is not None:
        try:
            train_loss = float(train_loss) if np.isfinite(float(train_loss)) else float('inf')
            eval_loss = float(eval_loss) if np.isfinite(float(eval_loss)) else float('inf')
            squad_f1 = float(squad_f1) if np.isfinite(float(squad_f1)) else 0
            squad_exact_match = float(squad_exact_match) if np.isfinite(float(squad_exact_match)) else 0
        except (TypeError, ValueError):
            train_loss = float('inf')
            eval_loss = float('inf')
            squad_f1 = 0
            squad_exact_match = 0
        # Store original run name or ID
        original_run_name = run.name  # Original name in W&B
        results.append({
            "rank": rank,
            "lr": lr,
            "modules": module_short_name,
            "train_loss": train_loss,
            "eval_loss": eval_loss,
            "train_loss_val_loss_gap": abs(train_loss - eval_loss),
            "trainable_params": trainable_params,
            "squad_exact_match": squad_exact_match,
            "squad_f1": squad_f1,
            "run_name": f"LoRA_r{rank}_lr{lr:.0e}_modules_{module_short_name}",  # Your new name
            "original_run_name": original_run_name  # Store original name
        })
```

## Statistical Analysis

```python
train_loss_stats = df["train_loss"].describe()
eval_loss_stats = df["eval_loss"].describe()
gap_stats = df["train_loss_val_loss_gap"].describe()

print("\nStatistical Analysis:")
print("Train Loss Stats:\n", train_loss_stats)
print("Eval Loss Stats:\n", eval_loss_stats)
print("Train-Val Loss Gap Stats:\n", gap_stats)
```

## Plot 1: All 75 results - Loss Curves

```python
plt.figure(figsize=(14, 8))
cmap = cm.get_cmap('tab20', len(results))  # Use distinct colors for 75 runs
for i, result in enumerate(results):
    plt.plot([1, 2, 3, 4, 5], [result["train_loss"]] * 5, color=cmap(i), 
             label=f"Train {result['run_name']}" if i % 2 == 0 else None)  # Label every other run for clarity
    plt.plot([1, 2, 3, 4, 5], [result["eval_loss"]] * 5, '--', color=cmap(i), 
             label=f"Eval {result['run_name']}" if i % 2 == 0 else None)
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training and Validation Loss Curves for All 75 Experiments")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10, ncol=2)
plt.tight_layout()
plt.savefig("all_loss_curves.png", dpi=300, bbox_inches='tight')
plt.show()
wandb.log({"all_loss_curves": wandb.Image("all_loss_curves.png")})
plt.close()
```

## Plot 2: All 75 results - Parameter Count vs. Performance (F1 and Loss)

```python
# Initialize W&B for plotting
wandb.init(project="lora_squad_finetuning", name="improved_scatter_plot")

# Set academic plotting style
plt.style.use('seaborn')
sns.set_context("paper", font_scale=1.2)

# Create scatter plot with dual Y-axes for all 75 results
fig, ax1 = plt.subplots(figsize=(12, 7))

# Color mapping for different modules
module_colors = {
    "attn": sns.color_palette("deep")[0],
    "attn_mlp": sns.color_palette("deep")[1],
    "attn_qv": sns.color_palette("deep")[2]
}

# Plot eval_loss on left Y-axis
for result in results:
    ax1.scatter(result["trainable_params"], result["eval_loss"], 
                c=module_colors[result["modules"]], s=50, alpha=0.6, 
                label=f"Eval Loss ({result['modules']})" if result == next(r for r in results if r["modules"] == result["modules"]) else None)

# Set up right Y-axis for squad_f1
ax2 = ax1.twinx()
for result in results:
    ax2.scatter(result["trainable_params"], result["squad_f1"], 
                c=module_colors[result["modules"]], marker='x', s=50, alpha=0.6, 
                label=f"SQuAD F1 ({result['modules']})" if result == next(r for r in results if r["modules"] == result["modules"]) else None)

# Annotate the best F1 score
best_f1_result = max(results, key=lambda x: x["squad_f1"])
ax2.annotate(f"Best F1: {best_f1_result['squad_f1']:.2f}\n{best_f1_result['run_name']}", 
             xy=(best_f1_result["trainable_params"], best_f1_result["squad_f1"]), 
             xytext=(best_f1_result["trainable_params"] + 10000, best_f1_result["squad_f1"] + 5),
             arrowprops=dict(facecolor='black', shrink=0.05), fontsize=10)

# Customize axes
ax1.set_xlabel("Number of Trainable Parameters", fontsize=12)
ax1.set_ylabel("Evaluation Loss", fontsize=12, color='blue')
ax2.set_ylabel("SQuAD F1 Score", fontsize=12, color='red')
ax1.tick_params(axis='y', labelcolor='blue')
ax2.tick_params(axis='y', labelcolor='red')
plt.title("Parameter Count vs. Performance for All 75 Experiments", fontsize=14, pad=15)

# Combine legends
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10)

# Adjust layout
plt.tight_layout()
plt.savefig("all_param_vs_performance_improved.png", dpi=300, bbox_inches='tight')
wandb.log({"all_param_vs_performance_improved": wandb.Image("all_param_vs_performance_improved.png")})
plt.show()
plt.close()

# Log data as table for interactivity in W&B
df_scatter = pd.DataFrame(results)[["run_name", "rank", "lr", "modules", "trainable_params", "eval_loss", "squad_f1"]]
df_scatter.columns = ["Run Name", "LoRA Rank", "Learning Rate", "Target Modules", "Trainable Parameters", "Eval Loss", "SQuAD F1"]
wandb.log({"scatter_data_table": wandb.Table(dataframe=df_scatter)})

# Finish W&B
wandb.finish()
```
